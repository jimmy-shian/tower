```python
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup

# 定義基本URL和分頁區間
base_url = "https://tos.fandom.com/zh/wiki/%E5%9C%96%E9%91%92/"
page_ranges = [
"1-300",
"301-600",
"601-900",
"901-1200",
"1201-1500",
"1501-1800",
"1801-2100",
"2101-2400",
"2401-2700",
"2701-3000",
"3001-3300",
"10001-10300",
"10301-10600",
"其它A-Z"
]

# 定義自定義函數，判斷元素是否符合條件
def has_custom_text(element):
    if element is None:
        return "None"
    return element.has_attr("class") and "boxtest" in element["class"] and "隊伍技能" in element.get_text()

# 創建一個空的 DataFrame
data = pd.DataFrame(columns=["Image Name", "Data Url", "Data Text", "Data Text2"]) # ["Image Name", "Table Data", "Table Text"]

# 迴圈處理不同分頁範圍
for i in page_ranges:
    page_url = f"{base_url}{i}"
    print(page_url)
    response = requests.get(page_url)
    print(response)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        images = soup.find_all("a", href=True, title=True)
        print('catch img')

        for image in images:
            image_name = image["title"]
            images_replace = image["href"].replace("/zh/wiki/", "")
            card_url = f"https://tos.fandom.com/zh/wiki/{images_replace}"
            card_response = requests.get(card_url)
            time.sleep(0.1)

            if card_response.status_code == 200:
                card_soup = BeautifulSoup(card_response.content, "html.parser")
                time.sleep(0.1)
                table = card_soup.select("table#monster-data")  # class="wikitable"   {"id": "monster-data"}
                table_text = table[0].prettify() if len(table) > 0 else ""

                if len(table) > 0:
                    print(f'find {image_name}')
                    elements = card_soup.find_all(has_custom_text)
                    elements_text = str(elements)
                    # 將資料添加到 DataFrame
                    data = data.append({"Image Name": image_name,
                                        "Data Url": card_url,
                                        "Data Text": table_text,
                                        "Data Text2": elements_text}, ignore_index=True)
                    # data = data.append([image_name, card_url, table_text, elements_text])
                    print(f'end {image_name}')
                    data.to_excel("monster_data_look.xlsx", index=False)
                    del image_name, card_url, table_text, elements_text
                    time.sleep(0.1)


# 將 DataFrame 寫入 Excel 檔案
data.to_excel("monster_data.xlsx", index=False)
print("Data saved to monster_data.xlsx")


# # 116m 41.8s

# # 67m 39.5s
# # 78m 33.4s

# # 267m 44.5s
```

# 合併sheets

```python
import pandas as pd

# 讀取 Excel 檔案
excel_file = r"C:\\Users\\Administrator\\Desktop\\tower\\monster_data.xlsx"
# excel_file = r"C:\\Users\\Administrator\\Desktop\\tower\\monster_data_look.xlsx"
xls = pd.ExcelFile(excel_file)

# 建立一個新的 DataFrame 用來存儲合併後的數據
merged_data = pd.DataFrame(columns=["Image Name", "Data Url", "Data Text", "Data Text2", "Sheet Name"])

# 遍歷每個分頁並合併數據
for sheet_name in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name)
    
    # 新增 "Sheet Name" 欄位，填入對應的分頁名稱
    df["Sheet Name"] = sheet_name
    # df = pd.concat([df, sheet_name])
    # 將該分頁的數據合併到總的 DataFrame
    merged_data = pd.concat([merged_data, df])

# 建立一個新的 Excel 檔案
new_excel_file = r"C:\\Users\\Administrator\\Desktop\\tower\\merged_monster_data.xlsx"
merged_data.to_excel(new_excel_file, index=False)
print("Merged data saved to", new_excel_file)
```

# 修正文字資料格式

```python
import pandas as pd
from bs4 import BeautifulSoup

# 讀取 Excel 檔案
excel_file = r"C:\\Users\\Administrator\\Desktop\\tower\\merged_monster_data.xlsx"
# excel_file = r"C:\\Users\\Administrator\\Desktop\\tower\\monster_data_look.xlsx"
parsed_data = pd.read_excel(excel_file)

# 建立一個空的 DataFrame 用來存儲新的 Excel 檔案資料
new_data = pd.DataFrame()
aa = 0
# 遍歷每一行資料
for index, row in parsed_data.iterrows():
    # 解析 "Data Text" 欄位的 HTML 內容
    soup_text = BeautifulSoup(row["Data Text"], "html.parser")
    aa+=1
    print(aa)

    image_link = soup_text.select('a', class_='image')[0]['href'].replace('amp;','')
    if aa == 1:
        print(image_link)
    if aa%10 == 0:
        print(image_link)
    # 使用 CSS 選擇器定位屬性和種族的元素
    attributes_element = soup_text.select_one("th:contains('屬性') + td")
    race_element = soup_text.select_one("th:contains('種族') + td")
    # 提取所需資料
    # 提取屬性和種族資訊
    attributes = attributes_element.get_text(strip=True) if attributes_element else "error"
    race = race_element.get_text(strip=True) if race_element else "error"

    # 使用 CSS 選擇器定位帶有特定文字的 <span> 元素
    min_cd_span = soup_text.select_one("span:contains('最小')")

    try:
        min_cd_span = soup_text.select_one("span:contains('最小')")
        min_cd = "最小"+ min_cd_span.get_text(strip=True).split("最小")[1] 

        min_cd_spans_element = soup_text.select("span:contains('最小')")
        # min_cd_content = min_cd_spans_element.find_next("td").get_text(strip=True)
        # print(min_cd_content)

        if len(min_cd_spans_element) >= 2:
            min_cd_contents = [element.find_next("td").get_text(strip=True) for element in min_cd_spans_element]  # 排除最後一個元素
        else:
            min_cd_content = min_cd_spans_element[0].find_next("td").get_text(strip=True)

    except:
        min_cd_contents = min_cd_content = "無"
        # min_cd = "技能CD"

    # 使用 CSS 選擇器定位主動技名稱和主動技效果的元素
    active_skill_name_elements = soup_text.select("th + td span.ability-tooltip a")
    # active_skill_effect_element = soup_text.select_one("th:contains('效果') + td")
    active_skill_effect_elements = soup_text.select("th:contains('效果') + td")
    # 
    # 判斷主動技名稱個數，如果大於2，則處理為多個主動技名稱，否則為一個主動技名稱
    if len(active_skill_name_elements) > 2:
        active_skill_names = [element.get_text(strip=True).replace(" ", "") for element in active_skill_name_elements[:-1]]  # 排除最後一個元素
    else:
        active_skill_name = active_skill_name_elements[0].get_text(strip=True).replace(" ", "") if active_skill_name_elements else "未找到主動技名稱"

    # 判斷主動技效果個數，如果大於2，則處理為多個主動技效果，否則為一個主動技效果
    if len(active_skill_effect_elements) > 2:
        active_skill_effects = [element.get_text(strip=True).replace(" ", "") for element in active_skill_effect_elements[:-1]]  # 排除最後一個元素
    else:
        active_skill_effect = active_skill_effect_elements[0].get_text(strip=True).replace(" ", "") if active_skill_effect_elements else "未找到主動技效果"

    # 使用 CSS 選擇器定位隊長技名稱和隊長技效果的元素
    leader_skill_name_element = soup_text.select_one("td:contains('隊長技') + th + td span.ability-tooltip a")
    leader_skill_effect_element = soup_text.select("th:contains('效果') + td")

    try:
        # 提取隊長技名稱和隊長技效果
        leader_skill_name = leader_skill_name_element.get_text(strip=True).replace(" ", "") if leader_skill_name_element else "未找到隊長技名稱"
        leader_skill_effect = leader_skill_effect_element[-1].get_text(strip=True).replace(" ", "") if leader_skill_effect_element else "未找到隊長技效果"
    except:
        leader_skill_name = "無"    
        leader_skill_effect = "無"

    data_text = {
        "圖片": image_link,
        "屬性": attributes,
        "種族": race,
        min_cd: min_cd_contents if len(min_cd_spans_element) > 1 else min_cd_content,
        "主動技名稱": active_skill_names if len(active_skill_name_elements) > 2 else active_skill_name,
        "主動技效果": active_skill_effects if len(active_skill_effect_elements) > 2 else active_skill_effect,
        "隊長技名稱": leader_skill_name,
        "隊長技效果": leader_skill_effect
    }

    # 解析 "Data Text2" 欄位的 HTML 內容
    soup_text2 = BeautifulSoup(row["Data Text2"], "html.parser")
    # 找到包含 "隊伍技能" 文字的 <div> 元素，提取隊伍技能的文字內容
    try:
        data_text2 = soup_text2.select_one("div:contains('隊伍技能')").get_text(strip=True).replace('隊伍技能','',1)
    except:
        data_text2 = "無隊伍技能"
    # data_text2 = soup_text2.get_text()

    # 將未處理的其他欄位資料也添加到新的資料字典中 # "Image Name", "Data Url", "Data Text", "Data Text2"
    new_data_row = {
        "Image Name": row["Image Name"],
        "Data Url": row["Data Url"],
        **data_text,
        # "Data Text": data_text,
        "Team Skills": data_text2
    }

    # 將整行資料添加到新的 DataFrame 中
    new_data = new_data.append(new_data_row, ignore_index=True)

    # try:
    #     del min_cd_content, min_cd_contents
    #     # del image_link, attributes, race, min_cd, min_cd_contents, min_cd_spans_element, min_cd_content, active_skill_names, active_skill_name_elements, active_skill_name, active_skill_effects, active_skill_effect_elements, active_skill_effect, leader_skill_name, leader_skill_effect, data_text2
    # except:
    #     # del min_cd_content
    #     pass 
        # del image_link, attributes, race, min_cd, min_cd_spans_element, min_cd_content, active_skill_name_elements, active_skill_name, active_skill_effect_elements, active_skill_effect, leader_skill_name, leader_skill_effect, data_text2
# 將新的資料寫入新的 Excel 檔案
new_excel_file = r"C:\\Users\\Administrator\\Desktop\\tower\\processed_monster_data.xlsx"
new_data.to_excel(new_excel_file, index=False)
print("Processed data saved to", new_excel_file)

# 4108
```

```python
import json
import re

# 读取原始的JSON数据
with open('new_final_data.json', 'r', encoding='utf-8') as json_file:
    original_data = json.load(json_file)

def process_text(text):
    processed_text = re.sub(r'\s+', ' ', re.sub(r'={3,}', '<br>================<br>', text))
    processed_text = processed_text.replace('：', '：<br><br>').replace(':', '：<br><br>').replace('：', '：<br><br>')  # Add <br><br> after ':'
    processed_text = processed_text.replace('◆', '<br>◆')  # Add <br> before '◆'
    processed_text = re.sub(r'={3,}', '<br>================<br>', processed_text)   
    processed_text = processed_text.replace("[", '').replace("\\", '').replace("'", '').replace("]", '').replace('xa0', '')  # [\'熾燄之傲‧強\', \'移靈加冕‧幽冥之燄\']  (包括「喋血屠刀」等扣除召喚師指定xa0%生命力的敵技所造成的傷害)

    # Add <br> before Roman numerals
    numerals = ['I.', 'II.', 'III.', 'IV.', 'V.', 'VI.', 'VII.', 'VIII.', 'IX.', 'X.']
    # for numeral in numerals: 
    #     if numeral in ['I.', 'II.', 'V.', 'X.']:
    #         pattern = f'(?<!I){re.escape(numeral)}'
    #         processed_text = re.sub(pattern, f'<br>{numeral} ', processed_text)
    #     else:
    #         processed_text = processed_text.replace(f'{numeral} ', f'<br>{numeral} ')
    numberss = ['⓵', '⓶', '⓷', '⓸', '⓹', '⓺', '⓻', '⓼', '⓽', '⓾' ]
    for ffg in numberss:
        processed_text = processed_text.replace(f'{ffg}', f'<br>{ffg}')

    for numeral in ['I.', 'IV.', 'VI.', 'VII.', 'VIII.', 'IX.']:
        pattern = f'(?<!I)(?<!V){re.escape(numeral)}'
        processed_text = re.sub(pattern, f'<br>{numeral} ', processed_text)
    for numeral1 in ['II.', 'V.', 'X.']:
        pattern2 = f'(?<!I){re.escape(numeral1)}'
        processed_text = re.sub(pattern2, f'<br>{numeral1} ', processed_text)
    for numeral2 in ['III.']:
        pattern3 = f'(?<!I){re.escape(numeral2)}'
        processed_text = re.sub(pattern3, f'<br>{numeral2} ', processed_text)
    
    # processed_text.replace(f'{numeral} ', f'<br>{numeral} ')
    # processed_text = re.sub(r'\s+', '', processed_text)
    processed_text = re.sub(r'(<br>){2,}', '<br><br> ', processed_text)
    return str(processed_text)

# 处理数据并生成HTML格式
html_data = []
for item in original_data:
    html_item = dict()
    try:
        min_ep_values = str(item["最小EP"])
        min_ep_values_str = min_ep_values.replace("'", "").replace("[", "").replace("]", "")  # 去掉字符串中的单引号
        min_ep_values = min_ep_values_str.split(", ")  # 分割字符串并生成列表
        min_ep_numbers = [int(value) for value in min_ep_values]  # 将列表中的字符串转换为整数
    except:
        min_cd_numbers = ['無']

    try:
        min_cd_values = str(item["最小CD"])
        min_cd_values_str = min_cd_values.replace("'", "").replace("[", "").replace("]", "")  # 去掉字符串中的单引号
        min_cd_values = min_cd_values_str.split(", ")  # 分割字符串并生成列表
        min_cd_numbers = [int(value) for value in min_cd_values]  # 将列表中的字符串转换为整数
    except:
        min_cd_numbers = ['無']
    # print(item)===============================================================
    if (str(item["最小EP"]) == 'None' and len(min_cd_numbers) == 1):

        active_skill = process_text(str(item["主動技名稱"])) + '：' + process_text(str(item["主動技效果"]))
        leader_skill = process_text(str(item["隊長技名稱"])) + '：' + process_text(str(item["隊長技效果"]))
        team_skill = process_text(str(item["Team Skills"]))

        html_item = {
    f'''
    <table class="table">
        <tr class="table-row-title">
            <th class="table-cell-title" style="width: 140px; overflow-y: hidden;">名稱</th>
            <th class="table-cell-title" style="width: 64px; overflow-y: hidden;">詳細介紹</th>
            <th class="table-cell-title" style="width: 50px; overflow-y: hidden;">縮圖</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">屬性</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">種族</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">最小CD</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">主動技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊長技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊伍技能</th>
        </tr>
        <tr class="table-row">
            <td class="table-cell" style="width: 140px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["Image Name"]}</td>
            <td class="table-cell" style="width: 64px; overflow-y: auto; align-items: center; justify-content: center;">
                <a href="{item["Data Url"]}" target="_blank">
                    點擊前往</a>
            </td>
            <td class="table-cell" style="width: 50px; overflow: hidden; align-items: center; justify-content: center;">
                <div class="image-wrapper">
                    <img src="{item["圖片"]}" alt="{item["Image Name"]}" width="40" height="auto">
                </div>
            </td>                
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["屬性"]}</td>      
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["種族"]}</td>
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["最小CD"]}</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {active_skill}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {leader_skill}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;"
            >{team_skill}<br>&nbsp</td>
        </tr>
    </table>  
    '''
        }
    if (str(item["最小EP"]) == 'None' and len(min_cd_numbers) > 1):

        active_skill_1 = str(item["主動技名稱"]).replace("'", "").replace("[", "").replace("]", "").split(", ")[0] + '：' + process_text(str(item["主動技效果"])).replace("'", "").replace("[", "").replace("]", "").split(", ")[0]
        active_skill_2 = str(item["主動技名稱"]).replace("'", "").replace("[", "").replace("]", "").split(", ")[1] + '：' + process_text(str(item["主動技效果"])).replace("'", "").replace("[", "").replace("]", "").split(", ")[1]
        leader_skill = str(item["隊長技名稱"]) + '：' + process_text(str(item["隊長技效果"]))
        team_skill = process_text(str(item["Team Skills"]))

        html_item = {
    f'''
    <table class="table">
        <tr class="table-row-title">
            <th class="table-cell-title" style="width: 140px; overflow-y: hidden;">名稱</th>
            <th class="table-cell-title" style="width: 64px; overflow-y: hidden;">詳細介紹</th>
            <th class="table-cell-title" style="width: 50px; overflow-y: hidden;">縮圖</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">屬性</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">種族</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">最小CD</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">主動技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊長技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊伍技能</th>
        </tr>
        <tr class="table-row">
            <td class="table-cell" style="width: 140px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["Image Name"]}</td>
            <td class="table-cell" style="width: 64px; overflow-y: auto; align-items: center; justify-content: center;">
                <a href="{item["Data Url"]}" target="_blank">
                    點擊前往</a>
            </td>
            <td class="table-cell" style="width: 50px; overflow: hidden; align-items: center; justify-content: center;">
                <div class="image-wrapper">
                    <img src="{item["圖片"]}" alt="{item["Image Name"]}" width="40" height="auto">
                </div>
            </td>                
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["屬性"]}</td>      
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["種族"]}</td>
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {min_cd_numbers[0]}</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {active_skill_1}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {leader_skill}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;"
            >{team_skill}<br>&nbsp</td>
        </tr>
        <tr class="table-row">
            <td class="table-cell" style="width: 140px; overflow-y: auto; align-items: center; justify-content: center;">
                </td>
            <td class="table-cell" style="width: 64px; overflow-y: auto; align-items: center; justify-content: center;">
                <a href="{item["Data Url"]}" target="_blank">
                    </a>
            </td>
            <td class="table-cell" style="width: 50px; overflow: hidden; align-items: center; justify-content: center;">
                <div class="image-wrapper">
                </div>
            </td>                
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                </td>      
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                </td>
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {min_cd_numbers[1]}</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {active_skill_2}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                <br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;"
            ><br>&nbsp</td>
        </tr>
    </table>  
    '''
        }
    if (str(item["最小CD"]) == 'None' and len(min_ep_numbers) == 1):

        active_skill = process_text(str(item["主動技名稱"])) + '：' + process_text(str(item["主動技效果"]))
        leader_skill = process_text(str(item["隊長技名稱"])) + '：' + process_text(str(item["隊長技效果"]))
        team_skill = process_text(str(item["Team Skills"]))

        html_item = {
    f'''
    <table class="table">
        <tr class="table-row-title">
            <th class="table-cell-title" style="width: 140px; overflow-y: hidden;">名稱</th>
            <th class="table-cell-title" style="width: 64px; overflow-y: hidden;">詳細介紹</th>
            <th class="table-cell-title" style="width: 50px; overflow-y: hidden;">縮圖</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">屬性</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">種族</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">最小CD</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">主動技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊長技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊伍技能</th>
        </tr>
        <tr class="table-row">
            <td class="table-cell" style="width: 140px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["Image Name"]}</td>
            <td class="table-cell" style="width: 64px; overflow-y: auto; align-items: center; justify-content: center;">
                <a href="{item["Data Url"]}" target="_blank">
                    點擊前往</a>
            </td>
            <td class="table-cell" style="width: 50px; overflow: hidden; align-items: center; justify-content: center;">
                <div class="image-wrapper">
                    <img src="{item["圖片"]}" alt="{item["Image Name"]}" width="40" height="auto">
                </div>
            </td>                
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["屬性"]}</td>      
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["種族"]}</td>
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["最小EP"]}</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {active_skill}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {leader_skill}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;"
            >{team_skill}<br>&nbsp</td>
        </tr>
    </table>  
    '''
        }
    if (str(item["最小CD"]) == 'None' and len(min_ep_numbers) > 1):

        active_skill_1 = str(item["主動技名稱"]).replace("'", "").replace("[", "").replace("]", "").split(", ")[0] + '：' + process_text(str(item["主動技效果"])).replace("'", "").replace("[", "").replace("]", "").split(", ")[0]
        active_skill_2 = str(item["主動技名稱"]).replace("'", "").replace("[", "").replace("]", "").split(", ")[1] + '：' + process_text(str(item["主動技效果"])).replace("'", "").replace("[", "").replace("]", "").split(", ")[1]
        leader_skill = str(item["隊長技名稱"]) + '：' + process_text(str(item["隊長技效果"]))
        team_skill = process_text(str(item["Team Skills"]))

        html_item = {
    f'''
    <table class="table">
        <tr class="table-row-title">
            <th class="table-cell-title" style="width: 140px; overflow-y: hidden;">名稱</th>
            <th class="table-cell-title" style="width: 64px; overflow-y: hidden;">詳細介紹</th>
            <th class="table-cell-title" style="width: 50px; overflow-y: hidden;">縮圖</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">屬性</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">種族</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">最小CD</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">主動技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊長技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊伍技能</th>
        </tr>
        <tr class="table-row">
            <td class="table-cell" style="width: 140px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["Image Name"]}</td>
            <td class="table-cell" style="width: 64px; overflow-y: auto; align-items: center; justify-content: center;">
                <a href="{item["Data Url"]}" target="_blank">
                    點擊前往</a>
            </td>
            <td class="table-cell" style="width: 50px; overflow: hidden; align-items: center; justify-content: center;">
                <div class="image-wrapper">
                    <img src="{item["圖片"]}" alt="{item["Image Name"]}" width="40" height="auto">
                </div>
            </td>                
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["屬性"]}</td>      
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["種族"]}</td>
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {min_ep_numbers[0]}</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {active_skill_1}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {leader_skill}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;"
            >{team_skill}<br>&nbsp</td>
        </tr>
        <tr class="table-row">
            <td class="table-cell" style="width: 140px; overflow-y: auto; align-items: center; justify-content: center;">
                </td>
            <td class="table-cell" style="width: 64px; overflow-y: auto; align-items: center; justify-content: center;">
                <a href="{item["Data Url"]}" target="_blank">
                    </a>
            </td>
            <td class="table-cell" style="width: 50px; overflow: hidden; align-items: center; justify-content: center;">
                <div class="image-wrapper">
                </div>
            </td>                
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                </td>      
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                </td>
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {min_ep_numbers[1]}</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {active_skill_2}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                <br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;"
            ><br>&nbsp</td>
        </tr>
    </table>  
    '''
        }
    if (min_cd_numbers == '無'):

        active_skill = process_text(str(item["主動技名稱"])) + '：' + process_text(str(item["主動技效果"]))
        leader_skill = process_text(str(item["隊長技名稱"])) + '：' + process_text(str(item["隊長技效果"]))
        team_skill = process_text(str(item["Team Skills"]))

        html_item = {
    f'''
    <table class="table">
        <tr class="table-row-title">
            <th class="table-cell-title" style="width: 140px; overflow-y: hidden;">名稱</th>
            <th class="table-cell-title" style="width: 64px; overflow-y: hidden;">詳細介紹</th>
            <th class="table-cell-title" style="width: 50px; overflow-y: hidden;">縮圖</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">屬性</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">種族</th>
            <th class="table-cell-title" style="width: 40px; overflow-y: hidden;">最小CD</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">主動技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊長技</th>
            <th class="table-cell-title" style="width: 175px; overflow-y: hidden;">隊伍技能</th>
        </tr>
        <tr class="table-row">
            <td class="table-cell" style="width: 140px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["Image Name"]}</td>
            <td class="table-cell" style="width: 64px; overflow-y: auto; align-items: center; justify-content: center;">
                <a href="{item["Data Url"]}" target="_blank">
                    點擊前往</a>
            </td>
            <td class="table-cell" style="width: 50px; overflow: hidden; align-items: center; justify-content: center;">
                <div class="image-wrapper">
                    <img src="{item["圖片"]}" alt="{item["Image Name"]}" width="40" height="auto">
                </div>
            </td>                
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["屬性"]}</td>      
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["種族"]}</td>
            <td class="table-cell" style="width: 40px; overflow-y: auto; align-items: center; justify-content: center;">
                {item["最小CD"]}</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {active_skill}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;">
                {leader_skill}<br>&nbsp</td>
            <td class="table-cell" style="width: 175px; overflow-y: auto; text-align: start ; justify-content: center;"
            >{team_skill}<br>&nbsp</td>
        </tr>
    </table>  
    '''
        }
    
    html_item = str(html_item).replace('\\n', '')
    print(html_item)
    html_data.append(html_item)
# 将处理后的HTML数据保存为新的JSON文件
with open('processed_data.json', 'w', encoding='utf-8') as json_file:
    json.dump(html_data, json_file, ensure_ascii=False, indent=4) 
```
